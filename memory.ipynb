```python
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_openai import ChatOpenAI


from langchain.globals import set_debug,set_verbose

set_verbose(True) # 没有更多的信息
set_debug(True)  # 全部的信息，太多了

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "尽你所能回答以下问题,控制在30个汉字"),
        ("placeholder", "{chat_history}"),
        ("human", "{input}"),
    ]
)

history = InMemoryChatMessageHistory()
llm = ChatOpenAI(model="gpt-4o-ca",)

chain = prompt | llm | StrOutputParser()

wrapped_chain = RunnableWithMessageHistory(chain, lambda x: history)

wrapped_chain.invoke(
    {"input": "我姐姐明天要过生日，我需要一束生日花束"},
    config={"configurable": {"session_id": "42"}},
)
```

```python
wrapped_chain.invoke(
    {"input": "她喜欢粉色玫瑰，颜色是粉色的。"},
    config={"configurable": {"session_id": "42"}},
)
```

```python
wrapped_chain.invoke(
    {"input": "我又来了，还记得我昨天为什么要来买花吗？"},
    config={"configurable": {"session_id": "42"}},
)
```
